# Features and Target

```python
 def rain(v):
        if v =='Yes':
            return 1
        return 0
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train, y_test = train_test_split(X,y)
```

```python
len(X_train)/len(df)
```

# Model Training

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

def return_scores(models, X_test, y_test):
    # models: a list of trained models# this function will return a list of accuracy, recall, precision, and f1 scores
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    
    # iterate through the models
    for m in models:
        predictions = m.predict(X_test)
        a = accuracy_score(y_test, predictions)
        p = precision_score(y_test, predictions)
        r = recall_score(y_test, predictions)
        f1 = f1_score(y_test, predictions)
        accuracy_scores.append(a)
        precision_scores.append(p)
        recall_scores.append(r)
        f1_scores.append(f1)

    #    Dictionary#d = {'acc': accuracy_scores,
    # 'pre': precision_scores,# 'rec': recall_scores,
    # 'f1': f1_scores}
    #return d
    return accuracy_scores, precision_scores, recall_scores, f1_scores
```

## Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
C_list = [10**e for e in range(-4,5)]
C_list
```

```python
#training 
lr_models = []

for c in C_list:
    lr = LogisticRegression(C=c,solver='liblinear')
    lr.fit(X_train,y_train)
    lr_models.append(lr)
    
acc, pre, rec, f1 =return_scores(lr_models,X_test,y_test) #unpacking tupple
```

```python
#visualization
acc, pre, rec, f1 =return_scores(lr_models,X_test,y_test) #unpacking tupple

plt.plot(C_list,acc, marker='o',label =' Test Accuracy')
plt.plot(C_list,pre, marker='o',label =' Test Precision')
plt.plot(C_list,rec, marker='o',label ='Test  Recall')
plt.plot(C_list,f1, marker='o',label =' Test F1')
plt.xlabel('C')
plt.legend()
plt.xscale('log')
plt.title('Model Results');
```

```python
#Balanced Training Data
lr_models_balanced = []

for c in C_list:
    lr = LogisticRegression(C=c,class_weight='balanced',solver='liblinear')
    lr.fit(X_train,y_train)
    lr_models_balanced.append(lr)
    
acc_bal, pre_bal, rec_bal, f1_bal =return_scores(lr_models_balanced,X_test,y_test) #unpacking tupple
```

```python
#visualization
plt.figure(figsize=(15,4))

ax1 = plt.subplot(1,2,1)
plt.plot(C_list,acc, marker='o',label =' Test Accuracy')
plt.plot(C_list,pre, marker='o',label =' Test Precision')
plt.plot(C_list,rec, marker='o',label ='Test  Recall')
plt.plot(C_list,f1, marker='o',label =' Test F1')
plt.xlabel('C')
plt.legend()
plt.xscale('log')
plt.title('Imbalanced')

plt.subplot(1,2,2,sharey=ax1)
plt.plot(C_list,acc_bal, marker='*',label =' Test Accuracy')
plt.plot(C_list,pre_bal, marker='*',label =' Test Precision')
plt.plot(C_list,rec_bal, marker='*',label ='Test  Recall')
plt.plot(C_list,f1_bal, marker='*',label =' Test F1')
plt.xlabel('C')
plt.legend()
plt.xscale('log')
plt.title('Balanced')

plt.suptitle('Logistic Regression Model Results')
plt.show();
```
